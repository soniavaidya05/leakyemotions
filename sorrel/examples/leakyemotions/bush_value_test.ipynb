```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ----------------------------------- #
# Some testing for dynamic bush value #
# ----------------------------------- #

RANGE = 30
LOC = RANGE / 2
SCALE = RANGE / 6
x = np.arange(0, RANGE+1)
plt.plot(x, norm.pdf(x, loc=LOC,scale=SCALE) / norm.pdf(x, loc=LOC,scale=SCALE).max() / 0.2, 'bo')
```
